## 01 – Dados as operações abaixo, qual o resultado da operação lógica:
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
#Instanciando operações
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

ol = p and q and r and s and t

#Resultado da operação lógica
print(f"O resultado da operação lógica (p and q and r and s and t) é: {ol}")

#Invertendo resultado
oli = p or q or r or s or t

#Resultado da operação lógica invertida
print(f"O resultado da operação lógica invertido (p or q or r or s or t) é: {oli}")
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
#Instanciando operação
p1 = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q1 = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r1 = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s1 = ['a',1,True,'b',0,False] > ['a',True,1]
t1 = ['a',1,True,'b',0,False] == 1
u1 = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

#Adicionando o operador `or`
op_or = p1 or q1 or r1 or s1 or t1 or u1

#Adicionando o operador `not or`
op_notor = not(p1 or not q1 or not r1 or not s1 or not t1 or not u1)

#Imprimindo resultado
print(f"O resultado da operação lógica (p1 or q1 or r1 or s1 or t1 or u1) é: {op_or}")
print(f"O resultado da operação lógica invertido (p1 or q1 or r1 or s1 or t1 or u1) é: {op_notor}")
```

## 03 – No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas. Importe aqui o mesmo objeto e arredonde e edite o index como realizado no exercício 3 da aula 8. Usando operadores responda, a segunda linha é igual a terceira ou a terceira linha é igual a quarta?

```python
#Importando pandas
import pandas as pd

#Importando tabela do exercício anterior
df = pd.read_csv('/content/tabela_ex07.csv', index_col= "Unnamed: 0")

#Arredondando valores
df = df.round(3)

#Operações
op_df = df.iloc[1].to_list() == df.iloc[2].to_list() or df.iloc[2].to_list() == df.iloc[3].to_list()

#Imprimindo resultado
print(f"A segunda linha é igual a terceira ou a terceira linha é igual a quarta -> {op_df}")
```

## 04 – Usando operadores verifique se todos os valores contidos na primeira linha são maiores ou iguais a 10.

```python
#Operação
linha_1 = (df.iloc[0] >=10).all()

#Imprimindo resultado
print(f"Todos os valores contidos na primeira linha são maiores ou iguais a 10 -> {linha_1}")
```

## 05 – Usando operadores verifique se todos os valores contidos na terceira coluna estão entre 0 inclusive e 40.

```python
#Operação
coluna_3 = (df.iloc[:,2] >= 0).all() & (df.iloc[:,2] <= 40).all()

#Imprimindo resultado
print(f"Todos os valores contidos na terceira coluna estão entre 0 inclusive e 40 -> {coluna_3}")
```

## 06 – Usando operadores verifique se algum dos valores contidos na última linha está acima de 10.

## 07 – Recupere a DataFrame contendo apenas valores que estejam acima de 5 e abaixo de 7.

## 08 – Recupere a DataFrame contendo apenas valores que estejam abaixo de 5 inclusive ou acima de 7.

```python
#Operações
df_5ou7 = df[(df < 5) | (df > 7)]

#Imprimindo resultado
df_5ou7
```

## 09 – Converta as duas últimas colunas do objeto DataFrame para dois objetos da classe numpy array. Algum dos valores contidos no primeiro array estão entre 2 inclusive e 5 inclusive?

```python
#Importando numpy
import numpy as np

#Localizando  colunas
col_1 = df.iloc[:,-1]
col_2 = df.iloc[:,-2]

#Convertendo em array
array_1 = np.array(col_1)
array_2 = np.array(col_2)

#Operação
ap_array = ((array_1 >= 2) & (array_1 <= 5)).any()

#Imprimindo resultado
print(f"Algum dos valores contidos no primeiro array estão entre 2 inclusive e 5 inclusive -> {ap_array}")

```
