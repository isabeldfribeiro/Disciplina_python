## 01 – Dados as operações abaixo, em Python, qual o resultado da operação lógica:
p and q and r and s and t?
<br>
<br>Imprima os objetos e o resultado usando uma estrutura condicional.

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
#Operações
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

#Imprimindo os objetos
print(f"{p=}")
print(f"{q=}")
print(f"{r=}")
print(f"{s=}")
print(f"{t=}")

#Operação lógica com estrutura condicional
resultado = p and q and r and s and t

#Condicional e resultado final
if resultado:
    print("Resultado da operação (p and q and r and s and t): True")
else:
    print("Resultado da operação (p and q and r and s and t): False")
```

## Instancie um objeto da classe array numpy, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R:. 1 -> Importaremos o Numpy logo que a função de números randômicos é desse pacote;

2 -> Deveremos chamar a função de números randômicos (random.normal()), onde daremos loc(média)=10, scale(desvio padrão)=3, e size(tamanho)=5, valores  esses exigidos pelo enunciado. Essa etapa é essencial para a obter os número randômicos e para que seja possível realizar as próximas etapas;

3 -> Será criada um lista para que possam ser armazenados os valores maiores que 10 e posteriormente possam ser contados;

4 -> Será iniciado um loop utilizando o "for", que é um elemento de repetição para que ele possa repetir as condicionais enquanto existirem valores randômicos;

5 -> Será utilizada a condicional para avaliar se os valores são maiores do que 10;

5.1 -> Caso o valor seja menor do que 10, e ainda restem valores randômicos, haverá um loop para o próximo valor, ignorando as instruções de dentro da condicional (i>10);

5.2 -> Caso o valor seja maior do que 10, o valor será adicionado a lista de armazenamento;

6 -> Os valores de dentro da lista de armazenamento serão exibidos;

7 -> Serão contados quantos números existem dentro da lista, e portanto, a quantidade de números maiores do que 10;

8 -> Será exibido a quantidade de valores maiores do que 10.

## 03 – Desenhe o fluxograma da resposta:

Fluxograma_valores_randômicos.png

```python
#Importando numpy
import numpy as np

#Instanciação dos valores randômicos
valores = np.random.normal(10, 3, 5)
print(f"Os valores gerados são: {valores}")

#Variavel para armazenamento de informações
maior_que_10 = []

#Condicional
for valor in valores:
    if valor > 10:
        maior_que_10.append(valor)
    else:
        pass

#Imprimindo resultados
print(f"Existem {len(maior_que_10)} valores maiores que 10 e são eles: {maior_que_10}")
```

```python
#Importando pandas
import pandas as pd

#Importando tabela do exercício 7 da aula 6 e colocando os genes como index
df = pd.read_csv('/content/genes.csv', index_col=0)

#Identificando primeiro gene
nome_gene_1 = df.index[0]
print(f"O primeiro gene é: {nome_gene_1}")

#Identificando número de caracteres do primeiro gene
num_caracteres_gene_1 = (df.iloc[0,0])
print(f"O gene {nome_gene_1} possui {num_caracteres_gene_1} caracteres")

#Condicional
if 2  <= num_caracteres_gene_1 <= 4:
    print(f"O gene {nome_gene_1} tem entre 2 e 4 caracteres")
    print("É possível identificar a informação gênica")
    print("As informações gênicas são:")
    print(df.loc[nome_gene_1])

elif num_caracteres_gene_1 <2:
    print(f"O gene {nome_gene_1} tem menos de 2 caracteres")
    print("Informações de todos os genes com mais de 4 caracteres:")
    print(df[df['n_caracteres']>4])

else:
    print(f"O gene {nome_gene_1} possui mais de 4 caracteres")
    print("Genes com mais de 4 caracteres")
    print("Informações de todos os genes com mais de 4 caracteres:")
    print(df[df['n_caracteres']>4].index.to_list())
```
