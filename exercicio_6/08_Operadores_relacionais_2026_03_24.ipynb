## 01 – Obtenha os resultados e retorne o código Python dos seguintes cálculos:

<br>a = 2.0 + 3.0 / 5.0 == 10
<br>b = (a + 10.2) - 5.0 * 2.0 <= 2.9
<br>c = 10 <= a / b + 33.8 < 100
<br>d = 5 != a / (b + c)
<br>e = 33 == d * 2

```python
#Instanciando expressões
a = 2.0 + 3.0 / 5.0 == 10
b = (a + 10.2) - 5.0 * 2.0 <= 2.9
c = 10 <= a / b + 33.8 < 100
d = 5 != a / (b + c)
e = 33 == d * 2

#Imprimindo resultados
print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")
print(f"d = {d}")
print(f"e = {e}")
```

## 02 – Altere os operadores relacionais para modificar o resultado de cada uma das comparações.

<br>('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]

<br>['a',1,True,'b',0,False] == 1
<br>['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

## 03 – Usando o método DataFrame.to_csv(‘nome_do_arquivo.csv’) exporte o objeto obtido no exercício 10 da aula anterior para um arquivo CSV. Usando o método pd.read_csv(‘nome_do_arquivo.csv’) instancie uma DataFrame. Use o valor ‘Unnamed: 0’ no parâmetro index_col, para ajustar o index. Arredonde os valores para até três casas decimais, usando o método round().
Recupere a DataFrame.

```python
#Importando pandas
import pandas as pd

#Importando tabela do exercício anterior
df = pd.read_csv('/content/tabela_ex07.csv', index_col= "Unnamed: 0")

#Arredondando valores
df = df.round(3)

#Recuperando df
df
```

## 04 – Instancie um numpy array que apresente o mesmo formato (linhas e colunas) da DataFrame. Este array deve conter apenas inteiros uns.
Recupere o array.

Obs.: Procure na internet o método numpy.ones()

```python
#Importando numpy
import numpy as np

#Obtendo o formato do df
sh = df.shape

#Criando array
array_1 = np.ones((sh))

#Recuperando array
array_1
```

## 05 – Quantos dos valores contidos na DataFrame são iguais aos valores no array?

```python
#Transformando o df em array
array_2 = np.array(df)
array_2

#Comparando arrays
array_igual = array_2 == array_1

#Número de valores iguais
valores = array_igual.sum()

#Imprimindo número de valores iguais
print(f"Os valores iguais são: {valores}")
```

## 06 – Quantos dos valores contidos na DataFrame são superiores aos valores no array?

```python
#Valores do array_1 em comparação ao array_2
array_maior = array_2 > array_1

#Número de valores maiores
valores_maior = array_maior.sum()

#Imprimindo número de valores maiores
print(f"O número valores do dataframe (array_2) maiores que os do array_1 são: {valores_maior}")
```

## 07 – Quantos dos valores contidos na DataFrame são iguais ou inferiores aos valores no array?

```python
#Valores do array_1 em comparação ao array_2
array_menor = array_2 <= array_1

#Número de valores menores
valores_menor = array_menor.sum()

#Imprimindo número de valores menores
print(f"O número valores do dataframe (array_2) menores ou iguais que os do array_1 são: {valores_menor}")
```

## 08 – Quantos dos valores contidos na DataFrame estão entre 0 e 1?

```python
#Valores do array_1 em comparação ao array_2
array_entre = (array_2 >= 0) & (array_2 <= 1)

#Número de valores entre 0 e 1
valores_entre = array_entre.sum()

#Imprimindo número de valores entre 0 e 1
print(f"O número valores do dataframe (array_2) entre 0 e 1 são: {valores_entre}")
```

## 09 – Quantos dos valores das médias das linhas são inferiores ou iguais à 5?

Obs 1.: Usando o método de classe, você pode converter qualquer objeto em outra classe. Ex. obj = list(array)

Obs 2.: Assim com o objeto da classe str, listas também possuem a função count()

```python
#Instanciando objeto com a última coluna(média das linhas)
obj = list(array_2[:, -1])

#Contando quantas médias menores ou iguais a 5 existem
contando = sum(obj.count(x) for x in set(obj) if x <= 5)

#Imprimindo resultado
print(f"{contando} valores são menores ou iguais a 5")
```

## 10 – Quantos dos valores das médias das colunas são superiores à 5?

```python
#Instanciando objeto com a última linha(média das colunas)
obj_1 = list(array_2[-1,:])

#Contando quantas médias menores ou iguais a 5 existem
contando_1 = sum(obj_1.count(x) for x in set(obj_1) if x > 5)

#Imprimindo resultado
print(f"{contando_1} valores são maiores ou iguais a 5")
```
