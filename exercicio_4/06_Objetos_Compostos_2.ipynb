## 1 – Um objeto composto multidimensional (matriz) pode receber  objeto composto unidimensional (vetor)?

Uma matriz pode receber um vetor.

```python
#instanciando uma matriz
matriz = [[1,2,3],[4,5,6],[7,8,9]]

#Imprimindo a matriz inicial
print(f"A matriz inicial é: {matriz}")

#Adicionando o vetor na matriz
matriz.append([10,11,12])

#Imprimindo a matriz ffinal
print(f"A matriz final é: {matriz}")
```

## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
#Importando tabela
df_seguro = pd.read_csv('data.csv')

#Número de entradas
n_ent = df_seguro.shape[0]

#Classe das colunas
colunas = list(df_seguro.dtypes)

#Número de elementos por classe
elem_classe = df_seguro.dtypes.value_counts().to_list()

#Determinando a média
med = df_seguro.iloc[:,[0,3,8]].mean().to_list()

#Determinando o desvio padrão
dp = df_seguro.iloc[:,[0,3,8]].std().to_list()

#Determinando o terceiro quartil
q3 = df_seguro.iloc[:,[0,3,8]].quantile(0.75).to_list()

#Criando dicionário
dicionario = {'Número de entradas': n_ent,
              'Classe das colunas': colunas,
              'Número de elementos por classe': elem_classe,
              'Média': med,
              'Desvio padrão': dp,
              'Terceiro quartil': q3}

#Imprimindo dicionário
dicionario
```
