## 01 – Obtenha os resultados e retorne o objeto e código em Python dos seguintes cálculos:
a) 2.0 + 3.0 / 5.0

b) (a + 10.2) - 5.0 * 2.0

c) a / b + 33.8

d) a / (b + c)

e) d * 2

```python
#Instanciando expressões
a = 2.0 + 3.0 / 5.0
b = (a + 10.2) - 5.0 * 2.0
c = a / b + 33.8
d = a / (b+c)
e = d * 2

#Imprimindo resultados
print(f'a = {a}')
print(f'b = {b}')
print(f'c = {c}')
print(f'd = {d}')
print(f'e = {e}')
```

## 02 – Obtenha os resultados e retorne o código do objeto em Python dos seguintes cálculos:
a) [2.0, 3.0, 5.0] + [10, 20, 30]

b) lista_a * 2

c) lista_a + lista_b

d) lista_a / (lista_b * lista_c)

```python
#Instanciando expressões
lista_a = [2.0, 3.0, 5.0] + [10, 20, 30]
lista_b = lista_a * 2
lista_c = lista_a + lista_b
#lista_d = lista_a / (lista_b * lista_c) -> Não é possível multiplicar uma lista por outra

#Imprimindo resultado das listas
print(f"lista_a = {lista_a}")
print(f"lista_b = {lista_b}")
print(f"lista_c = {lista_c}")
print(f"lista_d = Não é possível multiplica ou dividir uma lista por outra")
```

## 03 – Usando os objetos simples obtidos no exercício 01 (a, b, c, d, e), instancie uma  lista. Instancie também uma segunda lista contendo a raiz quadrada dos valores obtidos nos objetos simples. Instancie uma terceira lista, concatenando as duas primeiras. Recupere a última lista.

```python
#Instanciando lista
lista = [a, b, c, d, e]
lista_r2 = [a**0.5, b**0.5, c**0.5, d**0.5, e**0.5]
lista_final = lista + lista_r2

#Imprimindo a última lista
lista_final
```

## 04 – Duplique os valores contidos na lista final obtida no exercício anterior e converta para um array numpy. Remodele (reshape) o array, para que contenha 5 linhas e 4 colunas. Recupere o array.

```python
#Importando numpy
import numpy as np

#Duplicando lista_final
dup = lista_final*2

#Convertendo para np.array
array_dup = np.array(dup)

#Remodelando
array_dup = np.reshape(array_dup, (5,4))

#Imprimindo array
array_dup
```

## 05 – Do array numpy obtenha uma lista contendo a média dos valores de cada uma das colunas.

```python
lista_med = array_dup.mean(axis=0)
lista_med = lista_med.tolist()
lista_med
```

## 07 – Remodele (reshape) o array do exercício anterior, para que contenha 4 colunas. Converta este objeto da classe array para DataFrame. Arredonde todos os valores para até 3 casas decimais. Recupere a DataFrame.

```python
#Importando pandas
import pandas as pd

#Remodelando array -> Mesmo resultado do que array_med
array_col = array_med.reshape(-1,4)

#Convertendo para dataframe
df = pd.DataFrame(array_col)

#Arredondando valores
df = df.round(3)

#Imprimindo dataframe
df
```

## 08 – A última instancia desta DataFrame deve conter as médias dos valores contidos por coluna, o que foi determinado nos exercícios anteriores. Instancie um objeto contendo as médias dos valores em cada linha. Insira este objeto como uma nova coluna na DataFrame. Recupere a DataFrame.

```python
#Instanciando lista de média das linhas
lista_linhas_med = df.mean(axis=1)

#Adicionando no df
df[4] = lista_linhas_med

#Imprimindo df
df
```

## 09 – Usando o método dos objetos de classe DataFrame, describe(), compare os resultados contidos na última linha e última coluna da DataFrame.

Obs.: O método dos objetos de classe DataFrame, transpose(), inverte linhas por colunas.

```python
#Última coluna
desc_col = df.iloc[:, -1].describe()

#Transpondo df
transp = df.transpose()

#Última linha do df original
desc_lin = transp.iloc[:, -1].describe()

#Imprimindo resultado
print(desc_col)
print(desc_lin)
```

## 10 – Edite o nome das colunas e dos index, para indicar onde estão as médias. Qual a média das médias?

```python
#Editando nome das colunas
df = df.rename(columns = {4:'Médias'})

#Editando nome do index
df = df.rename(index = {5:'Médias'})

#Imprimindo df ajustado
print(df)

#Média das médias
med_med = df.iloc[-1,-1]

#Imprimindo média das médias
print(f"A média das médias é {med_med}")
```
