## 03 – Escreva o programa em Python:

```python
#Chamando pandas e matplotlib.pyplot
import pandas as pd
import matplotlib.pyplot as plt

#Chamando df
df = pd.read_csv('/content/genes.csv', index_col=0)

#Calculando frequência
freq = df["n_caracteres"].value_counts().sort_index()

#Inserindo no df
df["freq_caracteres"] = df["n_caracteres"].map(freq)

#Configurações
freq.plot(kind='bar')
plt.xlabel('Número de caracteres')
plt.ylabel('Frequência')
plt.title('Frequência do número de caracteres nos genes')
plt.xticks(rotation=0)
plt.show()
```

## 06 – Escreva o programa em Python:

```python
#Ordenando frequência
freq_2 = df["n_caracteres"].value_counts()
df["freq_caracteres"] = df["n_caracteres"].map(freq_2)

#Gerando gráfico de barras (usar dados agregados)
freq_2.sort_index().plot(kind='bar')

plt.xlabel('Número de caracteres')
plt.ylabel('Frequência')
plt.title('Frequência do número de caracteres nos genes')
plt.xticks(rotation=0)
plt.show()
```

## 07 – Reorganize a DataFrame, removendo as colunas contendo os dois primeiros caracteres, o número de caracter 'a', e a presença de 'hu', dos genes. Renomeie as colunas de acordo com as informações. Adicione uma coluna com a informação de 'Amostra 1'.

```python
#Removendo as colunas inicio_genes, número de a, e presença de hu
df_a1 = df.drop(columns=["inicio_genes","num_de_a","hu_no_nome"])

#Renomeando colunas
df_a1 = df_a1.rename(columns={'n_caracteres': 'size','freq_caracteres':'Amostra_1'})

#Imprimindo resultado
df_a1
```
