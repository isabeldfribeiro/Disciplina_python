# Exercício 04:

Para instanciar uma variável que recebe um valor inteiro, com o nome objeto_inteiro, como deve ser o código em Python?

# Exercício 10:

Dado a variável abaixo, qual método pode ser utilizado para buscar o número de ocorrências da sequencia “AUG” presentes no objeto?

<br>
<br>sequencia_gene = “ACTGGGUAUGAATTGCGCTTAGACTGGGUAUGAATTGCG”
<br>

```python
#Instanciando a sequencia
sequencia_gene = "ACTGGGUAUGAATTGCGCTTAGACTGGGUAUGAATTGCG"

#Contando a frequencia de 'AUG'
aug_seq = sequencia_gene.count("AUG")

#Imprimindo o resultado
print(f"A sequencia AUG aparece {aug_seq} vezes na sequencia")
```
