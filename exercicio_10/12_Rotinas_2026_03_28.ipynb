## Responda a pergunta do exercício 2 da aula 10, usando funções:<br>
Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

```python
#Importando pandas
import numpy as np

#Instancdo array
array = np.random.normal(10, 3, 5)
print(f"Os valores gerados são: {array}")

#Definindo da função
def maior_que_10(array):
    '''Avalia se os valores de uma array são maiores do que 10 e faz a contagem desses valores'''
    lista = []
    for i in array:
        if i > 10:
            lista.append(i)
    print(f"Os valores maiores que 10 são {lista}")
    print(f"{len(lista)} valores maiores que 10")

#Chamando função
maior_que_10(array)
```

## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

Seria possível, mas seria necessário fazer alterações na função criada para que ela seja aplicavel para valores maiores do que 20.

```python
#Importando pandas
import numpy as np

#Instancdo array
array = np.random.normal(10, 3, 5)
print(f"Os valores gerados são: {array}")

#Definindo da função
def maior_que_20(array):
    '''Avalia se os valores de uma array são maiores do que 20 e faz a contagem desses valores'''
    lista = []
    for i in array:
        if i > 20:
            lista.append(i)
    print(f"Os valores maiores que 20 são {lista}")
    print(f"{len(lista)} valores são maiores que 20")

#Chamando função
maior_que_20(array)
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
# Definindo a função
def divisão_obj_compostos(a, b):
    if type(a) not in [tuple, list] or type(b) not in [tuple, list]:
        print("Erro: Os objetos devem ser listas ou tuplas.")
        return

    #Validação de tamanho
    if len(a) != len(b):
        print("Erro: Os objetos devem ter o mesmo tamanho.")
        return

    #Divisão elemento a elemento
    resultado = []
    for i in range(len(a)):
        if b[i] == 0:
            print(f"Erro: Divisão por zero detectada no índice {i}.")
            return None

        divisao = a[i] / b[i]
        resultado.append(divisao)

    return resultado

#Testando
lista_1 = [10, 20, 30]
lista_2 = [2, 5, 10]

#Apllicando função
final = divisão_obj_compostos(lista_1, lista_2)
print(f"Resultado final: {final}")
```
