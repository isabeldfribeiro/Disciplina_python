## 01 – Quais as vantagens da utilização de objetos compostos em detrimento dos objetos simples?

A utilização de objetos compostos permite agrupar múltiplos objetos em uma única estrutura, facilitando a organização e manipulação do código. Seu uso também pode auxiliar na otimização no uso da memória RAM.

## 04 – Instancie uma variável da classe lista, contendo 5 nomes de genes diferentes, recupere de cada um dos valores quantas letras ‘a’ estão presentes:

```python
#Instanciando  lista de genes
genes = ["adeI","oxa-23","arr","sul","adeN"]

#Recuperando numero de letras "a"
gene_1 = genes[0].count("a")
gene_2 = genes[1].count("a")
gene_3 = genes[2].count("a")
gene_4 = genes[3].count("a")
gene_5 = genes[4].count("a")

#Imprimindo o número de letras "a" em cada gene
print(f"O gene {genes[0]} tem {gene_1} letras 'a'")
print(f"O gene {genes[1]} tem {gene_2} letras 'a'")
print(f"O gene {genes[2]} tem {gene_3} letras 'a'")
print(f"O gene {genes[3]} tem {gene_4} letras 'a'")
print(f"O gene {genes[4]} tem {gene_5} letras 'a'")
```

## 07 – Instancie uma variável da classe dicionário com quantos valores inteiros desejar. Usando as fun*ções globais, retorne a classe do objeto, quantos valores estão instanciad*os e a soma dos valores contidos no objeto:

```python
#Instanciando dicionario
dicionario_1 = {'a':75, 'b':8, 'c':54, 'd':29, 'e': 46 }

#Classe de objeto composto
dic_tipo = type(dicionario_1)

#Retornando a classe do dicionario_1
print(f"A classe do dicionario_1 é {dic_tipo}")

#Identificando numero de instanciacoes dentro de dicionario_1
dic_tamanho = len(dicionario_1)

#Imprimindo o numero de instanciacoes de dicionario_1
print(f"O dicionario_1 possui {dic_tamanho} instanciações")

#Somando os valores do dicionario
dic_soma = sum(dicionario_1.values())

#Imprimindo a soma dos valores inteiros do dicionario
print(f"A soma dos valores do dicionario_1 é {dic_soma}")

```

## 09 – Instancie cinco variáveis da classe lista. Cada uma delas será referente à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Em cada uma das listas, você deve inserir o número de caracteres presentes no nome do gene, os dois primeiros caracteres do nome, quantas vezes aparece o caractere ‘a’ e se no nome aparecem os caracteres ‘hu’. Para isso você deve utilizar a indexação de lista. Retorne cada um dos objetos:

```python
#Criando lista do primeiro gene
gene_1 = []

#Contando caracteres, identificando as duas primeiras letras, contando o número de 'a', e verificando a presença de 'hu'
gene_1.append(len(genes[0]))
gene_1.append(genes[0][0:2])
gene_1.append(genes[0].count('a'))
gene_1.append('hu' in genes[0])

#Imprimindo lista do primeiro gene
print(f"A lista do gene {genes[0]} é: {gene_1}")


#Criando lista do segundo gene
gene_2 = []

#Contando caracteres, identificando as duas primeiras letras, contando o número de 'a', e verificando a presença de 'hu'
gene_2.append(len(genes[1]))
gene_2.append(genes[1][0:2])
gene_2.append(genes[1].count('a'))
gene_2.append('hu' in genes[1])

#Imprimindo lista do segundo gene
print(f"A lista do gene {genes[1]} é: {gene_2}")


#Criando lista do terceiro gene
gene_3 = []
gene_3.append(len(genes[2]))
gene_3.append(genes[2][0:2])
gene_3.append(genes[2].count('a'))
gene_3.append('hu' in genes[2])

#Imprimindo lista do terceiro gene
print(f"A lista do gene {genes[2]} é: {gene_3}")


#Criando lista do quarto
gene_4 = []
gene_4.append(len(genes[3]))
gene_4.append(genes[3][0:2])
gene_4.append(genes[3].count('a'))
gene_4.append('hu' in genes[3])

#Imprimindo lista do quarto gene
print(f"A lista do gene {genes[3]} é: {gene_4}")


#Criando lista do quinto
gene_5 = []
gene_5.append(len(genes[4]))
gene_5.append(genes[4][0:2])
gene_5.append(genes[4].count('a'))
gene_5.append('hu' in genes[4])

#Imprimindo lista do quinto gene
print(f"A lista do gene {genes[4]} é: {gene_5}")
```
