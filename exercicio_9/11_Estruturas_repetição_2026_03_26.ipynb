## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0.
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: 1 -> Será instanciado o objeto de i=0 para que seja possível fazer o loop de 5000 valores;

2 -> Serão instanciadas 3 listas, uma para armazenar os valores do loop de 5000 valores, uma para armazenar os números pares e outro para números ímpares;

3 -> Será avaliado se o i é menor do que 5000;

3.1 -> Se o i for menor do que 5000, será adicionado mais 1 ao valor do i e posteriormente o valor do inicio do loop será armazenado na lista de armazenamento de 5000 valores.

3.2 -> Se i for maior que 5000 o loop se encerra e passa para a próxima etapa.

4 -> Para cada valor da lista de armazenamento de 5000 valores, será avaliado se o resto de uma divisão por dois será igual a 0;

4.1.1 -> Se sim, o valor será par, e deverá ser armazenado na lista de números pares;

4.1.2 -> Será contado a quantidade de números pares presentes na lista e esse número será exibido.

4.2.1 -> Se houver resto da divisão por 2, entende-se que o  valor é ímpar, e deverá ser armazenado na lista de números ímpares;

4.2.2 -> Será contado a quantidade de números ímpares presentes na lista e esse número será exibido.

## 02 – Desenhe o fluxograma da resposta:

Fluxograma_5000.png

## 03 – Escreva o programa em Python:

```python
#Instanciando lista vazia
lista = []

#Instanciando lista de 5000 valores a partir de 0
i = 0
while i < 5000:
    lista.append(i)
    i += 1
print(f"A lista instanciada é: {lista}")

#Identificando pares e immpares
pares = []
impares = []
for i in lista:
  if i % 2 == 0:
    pares.append(i)
  else:
    impares.append(i)
print(f"Existem {len(pares)} números pares")
print(f"Existem {len(impares)} números ímpares")
```

## Importe o .csv exportado na última aula e instancie um objeto da classe DataFrame.

Dos números de letras nos genes, quais são pares e quantos são impares?

## 05 – Descreva o racional da sua resposta:

R.: 1 -> Serão instanciadas 2 listas, uma para valores pares, e outra para valores ímpares;

2-> Será instanciado um valor de i (i = 0) para que seja poossível fazer o loop;

3 -> Será avaliado se o valor de i é menor do que o número de linhas do df;

3.1 -> Se sim se iniciará o loop.

3.2 -> Se o i for maior que o número de linhas do df, será impresso o resultado final.

4 -> Para cada linha do df será identificado o gene e seu respectivo número de caracteres;

5 -> Será avaliado se o resto da divisão do número de caracteres por 2 é igual a 0;

5.1 -> Se o resto for igual a 0, o gene será depositado na lista de números pares;

5.2 -> Se houver resto na divisão por 2, o gene será depositado na lista de números ímpares;

6 -> Após os genes serem depositados em suas respectivas listas, será adicionado mais 1 ao valor do i;

7 -> O i voltará a ser verificado, se possuí valor menor do que o número de linhas do df;

8-> Ao finalizar o loop serão exibidos quantos e quais são os genes ímpares e pares.

## 06 – Desenhe o fluxograma da resposta:

Fluxograma_n_caracteres_genes.png

## 07 – Escreva o programa em Python:

```python
#Importando pandas
import pandas as pd

#Importando df
df = pd.read_csv("genes.csv", index_col=0)

#Instanciando listas
pares_1 = []
impares_1 = []

#Instanciando valor de i
i = 0

#Estrutura de repetição
while i < len(df):
    carac = df['n_caracteres'].iloc[i]
    gene = df.index[i]
    if carac % 2 == 0:
        pares_1.append(gene)
        print(f"O gene {gene} possui um número par de caracteres ({carac})")
    else:
        impares_1.append(gene)
        print(f"O gene {gene} possui um número ímpar de caracteres ({carac})")

    #Para continuar o loop até cerrto ponto
    i += 1

print(f"Existem {len(pares_1)} genes com número par de caracteres, sendo eles {pares_1}")
print(f"Existem {len(impares_1)} genes com números ímpares de caracteres, sendo eles {impares_1}")
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.
